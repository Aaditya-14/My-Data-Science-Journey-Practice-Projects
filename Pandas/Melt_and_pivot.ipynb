```python
import pandas as pd

# Load the dataset
data = {
    'Student': ['Ram', 'Hari', 'Sita', 'Gita'],
    'Math': [80, 60, 90, 85],
    'Science': [75, 68, 95, 89],
    'English': [85, 70, 88, 82],
    'History': [70, 72, 85, 80]
}
df = pd.DataFrame(data)
```

```python
df
```

```python
d2=df.melt(id_vars=["Student"], 
        value_vars=["Math", "Science", "English", "History"], 
        var_name="Subject", 
        value_name="Score")
```

```python
d2
```

```python
#pivot is opposite of melt
d2.pivot(index="Student",
        columns="Subject",
        values="Score")
```
