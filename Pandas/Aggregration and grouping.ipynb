```python
import pandas as pd
```

```python
data=pd.read_csv("sales_data.csv")
```

```python
data
```

```python
#groupby Function
#df.groupby("Team")["Salary"].mean()
#Common Aggregation Functions
data.groupby("SalesPerson")["Revenue"].mean()
```

```python
data.groupby("SalesPerson")["Revenue"].sum()
```

```python
data.groupby("SalesPerson")["Revenue"].min()
```

```python
data.groupby("SalesPerson")["Revenue"].max()
```

```python
data.groupby("SalesPerson")["Revenue"].std()
```

```python
#using.agg help to find operation in group
data.groupby("SalesPerson")["Revenue"].agg(["min","max","mean","std","sum"])
```

```python
#put agg in table
data.groupby("SalesPerson")["Revenue"].agg(
    minimum_values="min",
    Maximum_values="max",
    Mean__values="mean",
    Standard_deviation="std",
    Sum_of_all_values="sum")
```

```python
#example of Transform
```

```python
data["Average"]=data.groupby("SalesPerson")["Revenue"].transform("mean")
```
