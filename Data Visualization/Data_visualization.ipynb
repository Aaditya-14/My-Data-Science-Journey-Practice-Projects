```python
import matplotlib.pyplot as plt
```

```python
plt.plot([1, 2, 3], [4, 5, 6])
```

```python
years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
runs =  [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]
plt.plot(years,runs)
plt.xlabel("Years")
plt.ylabel("runs")
plt.title("Peformance comparision of year and run")
```

```python
kohli = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]
sehwag = [0, 300, 800, 1200, 1500, 1700, 1600, 1400, 1000, 0]
```

```python
plt.plot(years,kohli,color="red",linewidth=3,linestyle='--',label="Kolhi run")
plt.plot(years, sehwag,color="Pink",linewidth=4,linestyle='-.', label="Virender Sehwag")
plt.xlabel("Years")
plt.ylabel("Years")
plt.title("Kohli and sehwag runs")
plt.tight_layout()
plt.legend()
plt.grid(True)
```
