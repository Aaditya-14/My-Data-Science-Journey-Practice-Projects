```python
import matplotlib.pyplot as plt
```

```python
study_hours = [1, 2, 3, 4, 5, 6, 7, 8, 9]
exam_scores = [40, 45, 50, 55, 60, 65, 75, 85, 90]
```

```python
for i in range(len(study_hours)):
    plt.annotate(f"s{(i+1)}",(study_hours[i]+0.1,exam_scores[i]-3))
color = ["green" if score < 60 else "red" for score in exam_scores]
size=[score * 2 for score in exam_scores]
plt.scatter(study_hours,exam_scores, c=exam_scores, s=size,cmap='viridis')
plt.colorbar(label='Score')
plt.xlabel("Marks")
plt.ylabel("Hours")
plt.title("Comparision of study hours and scores")
plt.grid(True)
```

```python
class_a_hours = [2, 4, 6, 8]
class_a_scores = [45, 55, 65, 85]
 
class_b_hours = [1, 3, 5, 7, 9]
class_b_scores = [40, 50, 60, 70, 90]
```

```python
for i in range (len(class_a_hours)):
    plt.annotate(f"a{i+1}",(class_a_hours[i]+0.1,class_a_scores[i]))
for j in range (len(class_b_hours)):
    plt.annotate(f"b{j+1}",(class_b_hours[j]+0.1,class_b_scores[j]-0.2))

plt.scatter(class_a_hours,class_a_scores, label="Class a score",c=class_a_scores,cmap='viridis')
plt.scatter(class_b_hours,class_b_scores,label="Class a score",c=class_b_scores,cmap='viridis')
plt.colorbar(label='Score')
plt.legend()
plt.title("Scores vs Hours Comparision")
plt.xlabel("Hours")
plt.ylabel("Scores")
plt.grid(True)
```
