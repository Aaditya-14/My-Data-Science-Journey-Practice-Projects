```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
x1=[i*2 for i in x]
y1=[i**2 for i in x]
```

```python
plt.subplot(2,2,1)
plt.plot(x,x1)
plt.title("Double plot")
plt.xlabel("x")
plt.ylabel("x1")
plt.tight_layout()

plt.subplot(2,2,2)
plt.plot(x,y1)
plt.title("Square Plot")
plt.tight_layout()
plt.xlabel("x")
plt.ylabel("y1")

plt.subplot(2,2,3)
plt.bar(x,y1)
plt.title("Double Bar")
plt.tight_layout()
plt.xlabel("x")
plt.ylabel("x1")

plt.subplot(2,2,4)
plt.bar(x,x1)
plt.title("Square Bar")
plt.tight_layout()
plt.xlabel("x")
plt.ylabel("y1")
```

```python
fig,axis=plt.subplots(2,2,figsize=(5,5))
axis[0,0].plot(x,x1)
axis[0,0].set_title("Plot1")
plt.tight_layout()

axis[1,0].plot(x,y1)
axis[1,0].set_title("Plot2")
plt.tight_layout()

axis[0,1].bar(x,x1)
axis[0,1].set_title("Plot3")
plt.tight_layout()

axis[1,1].bar(x,y1)
axis[1,1].set_title("Plot4")
plt.tight_layout()

fig.suptitle("4 Figure Comparision", fontsize="12")
plt.tight_layout()
fig.savefig('my_plots.png')
```
