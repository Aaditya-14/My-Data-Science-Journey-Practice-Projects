```python
import matplotlib.pyplot as plt
import numpy as np
```

```python

years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]

sachin = [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]
sehwag = [0, 200, 900, 1400, 1600, 1800, 1500, 1100, 800, 0]
kohli  = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]
```

```python
x=np.arange(len(years))
```

```python
width=0.30
```

```python
plt.bar(x-width,sachin,width=width, color="Red",label="sachin")
plt.bar(x,sehwag,width=width, color="Blue",label="sehwag")
plt.bar(x+width,kohli,width=width, color="Green",label="kohli")
plt.xlabel("Year")
plt.ylabel("Runs")
plt.title("Runs and Year comparision of Cricketer")
plt.grid(True)
plt.legend()

plt.xticks(x,years)
plt.tight_layout()
```

```python
players = ["Sachin", "Sehwag", "Kohli", "Yuvraj"]
run = [500+700+1100+1500+1800, 0+200+900+1400+1600, 0+0+500+800+1100, 300+600+800+1100+900]
```

```python
colour=['Red','green','pink','yellow']
plt.barh(players,run,color=colour,label=players)
plt.legend()
plt.xlabel("Year")
plt.ylabel("Runs")
plt.title("Runs and Year comparision of Cricketer")
plt.grid(True)

for i in range(len(players)):
    plt.text(run[i] + 50, i, str(run[i]), va='center')
```
